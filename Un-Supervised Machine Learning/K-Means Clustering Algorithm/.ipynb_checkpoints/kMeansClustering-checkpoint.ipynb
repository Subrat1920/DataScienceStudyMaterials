# K Means Clusetring

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
x, y = make_blobs(n_samples=1000, centers=3, n_features=2)
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
## standardization 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=42)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.cluster import KMeans
```

```python
import warnings as war
war.filterwarnings('ignore')
```

```python

## Elbow Method
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(x_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## Elbow Curve Plot
plt.plot(range(1,11), wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
```

```python
y_pred_train = kmeans.fit_predict(x_train_scaled)
y_pred_test = kmeans.fit_predict(x_test_scaled)
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c=y_pred_test)
```

```python
## for validataing k values
### 1. knee locaotor
### 2. Silhoutte scoring
```

### Knee Locator

```python
from kneed import KneeLocator
```

```python
k1=KneeLocator(range(1,11), wcss, curve='convex', direction='decreasing')
```

```python
k1.elbow
```

### Silhoutee Scoring

```python
from sklearn.metrics import silhouette_score
```

```python
silhoutee_coefficients = []
for k in range(2,11):
    kmeans=KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(x_train_scaled)
    score = silhouette_score(x_train_scaled, kmeans.labels_)
    silhoutee_coefficients.append(score)
```

```python
silhoutee_coefficients
```

```python
## Elbow Curve Plot
plt.plot(range(2,11), silhoutee_coefficients)
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhoutee Co-efficients')
```
