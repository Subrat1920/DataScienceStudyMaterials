## Algerian Forest Fires Dataset

Data Set Information

The Dataset includes 244 instances that group a data of two regions of Algeria, namely Bejala region located in the northeast Algeria and the Sisi Bel-Abbes region located in the northwest of the Algeria.

122 Instances for each region.

The Period from June 2012 to September 2012, The dataset includes 11 attributes and 1 outputs attribute (class). The 244 instances have been classified into fire (138 classes) and not fire (106 classes) not fire.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_column', None)
```

```python
adf = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv')
print(adf.shape)
adf.head()
```

```python
bdf = pd.read_csv('Bejaia Region ForestFire Dataset.csv')
bdf.head()
bdf.loc[0:]
```

```python
fdf = pd.concat([adf, bdf])
fdf.head()
```

```python
fdf.index
```

```python
fdf = fdf.reset_index(drop = True)
```

```python
fdf['Region'] = 0
fdf.loc[122:, 'Region'] = 1
```

### Exploratory Data Analysis

```python
data = fdf
```

```python
data.head()
```

```python
data['Classes  '].value_counts()
```

```python
data.columns
```

```python
## Encoding of the categories in classes

data['Classes  ']=np.where(data['Classes  '].str.contains('not fire'), 0, 1)
```

```python
data['Classes  '].value_counts()
```

### Density Plot for all Features

```python
plt.style.available
```

```python
plt.style.use('ggplot')
data.hist(bins=50, figsize=(20,15), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
## Percentage for Pie Chart
perc = data['Classes  '].value_counts(normalize=True)*100
```

```python
# Plotting Pie Charts
class_labels = ['Fire', 'Not Fire']
plt.figure(figsize=(10,5))
plt.pie(perc, labels=class_labels, autopct='%1.1f%%')
plt.title('Pie Chart for Classes')
plt.show()
```

```python
# Correlation
data.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
# Box Plots
sns.boxplot(data=data, x=data.FWI)
```
