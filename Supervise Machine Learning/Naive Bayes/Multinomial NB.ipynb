# Multinomial Naive Bayes

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
emails = [
    'Win a brand new car by clicking this link',
    'You account has been hacked, reset your password',
    'Limited time offer, Buy Now!!',
    'Meeting at 3, join now!!',
    'Your Subscription is about expire, recharge now.',
    'Click and Get 500000 voucher for free!!',
    'Yor interview is scheduled at 12 tommorow.'
]

# 1 ===> for Spam emails
# 0 ===> for non Spam emails

labels = [1,1,1,0,0,1,0] 
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(emails)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,labels, test_size=0.3, random_state=42)
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train, y_train)
```

```python
y_pred = mnb.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
text = 'Grab opportunity to get free car'
tr_text = vectorizer.fit_transform(text)
mnb.predict([[tr_text]])
```
