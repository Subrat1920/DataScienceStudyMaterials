```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
print('sucessfull')
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
## taking only two categorical species values
df = data[data['species']!='setosa']
df.species.value_counts()
```

```python
## splitting the data
x = df.drop(columns='species', axis=1)
y = df.species
```

```python
## changing the string values to numerical values
y = y.map({
    'versicolor':0, # for versicolor value is 0
    'virginica' :1  # for verginica value is 1
})
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import xgboost as xgb
model = xgb.XGBClassifier()
```

```python
params = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=params, cv=5)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
best_model = grid.best_estimator_
best_model
```

```python
y_pred_train = best_model.predict(x_train)
y_pred_test = best_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred_train, y_train), accuracy_score(y_pred_test, y_test)
```
