# Named Entity Recognition
##### Named Entity Recognition (NER) is a subfield of Natural Language Processing (NLP) that involves identifying and categorizing specific data points, known as named entities, within unstructured text. These named entities are single words, phrases, or sequences of words that represent predefined categories, such as:
    Person names: Individuals, including celebrities, politicians, and ordinary people
    Organizational names: Companies, institutions, governments, and non-profit organizations
    Geographic locations: Cities, countries, regions, and landmarks
    Dates: Specific days, months, years, and time periods
    Events: Historical events, conferences, and meetings
    Quantitative values: Monetary amounts, percentages, and numerical expressions

```python
sentence = "My name is Subrat Mishra, and I am an aspiring Data Scientist"
```

```python
import nltk
nltk.word_tokenize
```

```python
tokens = nltk.word_tokenize(sentence)
tagged_elements = nltk.pos_tag(tokens)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
```

```python
nltk.ne_chunk(tagged_elements).draw()
```
