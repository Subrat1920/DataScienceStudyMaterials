```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
texts = [
    'There is Horse',
    'I am a Student',
    'I am curious for Software Development',
    'I love to read Data Science related articles',
    'I have a laptop'
]
```

```python
voc_size = 10000
```

```python
one_hot_rep = [one_hot(words, voc_size) for words in texts]
one_hot_rep
```

```python
max_length = 10
```

```python
## embedding the layer
embedded_words = pad_sequences(sequences=one_hot_rep, maxlen=max_length, padding='post')
embedded_words
```

```python
dim = 20
model = Sequential()
embedding_layer = Embedding(input_dim=voc_size, output_dim=dim)
model.add(embedding_layer)
embedding_layer.build((None, max_length))
```

```python
model.summary()
```

```python
model.predict(embedded_words)
```
