```python
import seaborn as sns
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
x = df.drop(columns=['species'], axis=1)
y = df[['species']]
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
scaler=StandardScaler()
le=LabelEncoder()
x_scaled = scaler.fit_transform(x)
y_encoded = le.fit_transform(y)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x_scaled,y_encoded,test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
x.shape[1]
```

```python
hidden_layers = [1,2,3,4]
neurons_per_layer = [32, 64, 128, 256]
activation_functions = ['relu', 'sigmoid']
learning_rate = [0.001, 0.01, 0.1]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

best_accuracy = 0
best_params = {}

for hid_layers in hidden_layers:
    for neurons in neurons_per_layer:
        for act_func in activation_functions:
            for lear_rate in learning_rate:
                print(f'Training with\n {hid_layers} [Hidden Layers]\n {neurons} [Neurons Per Layer]\n {act_func} [Activation Function]\n {lear_rate} [Learning Rate]')
                
                model = Sequential()  # Build the model
                model.add(Dense(neurons, activation=act_func, input_dim=x.shape[1]))  # Input layer

                for _ in range(hid_layers):
                    model.add(Dense(neurons, activation=act_func))  # Hidden layers
                
                model.add(Dense(3, activation='softmax'))  # Output layer

                opt = tf.keras.optimizers.Adam(learning_rate=lear_rate)  # Optimizer
                loss = tf.keras.losses.SparseCategoricalCrossentropy()  # Loss function

                # Compile the model
                model.compile(optimizer=opt, loss=loss, metrics=['accuracy'])

                # Train the model
                history = model.fit(x_train, y_train, batch_size=5, epochs=50, validation_split=0.2, verbose=2)

                val_accuracy = max(history.history['val_accuracy'])
                print(f'Validation Accuracy=====> {val_accuracy:.4f}')

                # Correct update of best accuracy and best parameters
                if val_accuracy > best_accuracy:
                    best_accuracy = val_accuracy  # Update the best accuracy
                    best_params = {  # Store best parameters
                        'hidden_layers': hid_layers,
                        'neurons_per_layer': neurons,
                        'activation_functions': act_func,
                        'learning_rate': lear_rate
                    }
```

```python
print("Best Hyperparameters:", best_params)
print("Best Validation Accuracy:", best_accuracy)
```

```python
best_params
```

```python
## creating the best model
best_model = Sequential([
    Dense(32, activation='relu', input_dim=x.shape[1]),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax'),
])
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.1)                                 ## added optimizer
loss = tf.keras.losses.SparseCategoricalCrossentropy()                            ## loss function
```

```python
best_model.compile(optimizer=opt, loss=loss, metrics=['accuracy'])
```

```python
best_model.fit(x_train, y_train, batch_size=5, epochs=50)
```

```python
loss, acc = best_model.evaluate(x_test, y_test)
loss, acc
```

```python
prediction = np.argmax(best_model.predict(x_test),axis=1)
```

```python
from sklearn.metrics import classification_report
classification_report(prediction, y_test)
```
